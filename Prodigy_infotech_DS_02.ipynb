```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df
```

```python
#Dropping unnecessary columns
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df
```

```python
df1['Sex'] = df1['Sex'].map({'male': 0, 'female': 1})
df1
```

```python
# Visualize the relationships between variables in test data
sns.pairplot(df1)
plt.show()
```

```python
sns.pairplot(df, hue='Survived')
plt.show()
```

```python
#the survival rate by different features
print(df.groupby('Pclass')['Survived'].mean())
print(df.groupby('Sex')['Survived'].mean())
print(df.groupby('Embarked')['Survived'].mean())
```

```python
#age distribution among survivors and non-survivors
plt.figure(figsize=(10, 5))
sns.histplot(df[df['Survived'] == 1]['Age'], label='Survived', kde=True)
sns.histplot(df[df['Survived'] == 0]['Age'], label='Did not survive', kde=True)
plt.legend()
plt.show()
```

```python
#fare distribution in test data
plt.figure(figsize=(10, 3))
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution in Test Data')
plt.show()
```

```python
#'Survived' distribution in gender submission
sns.countplot(data=gs, x='Survived')
plt.show()
```
